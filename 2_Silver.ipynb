# Silver Notebook Lookup Table


## **_Parameters_**

```python
dbutils.widgets.text("sourcefolder","netflix_directors")
dbutils.widgets.text("destinationfolder","netflix_directors")
```

```python
var_src_folder=dbutils.widgets.get('sourcefolder')
var_target_folder=dbutils.widgets.get('destinationfolder')
```

```python
df=spark.read.format('csv')\
    .option('header',True)\
        .option('inferSchema',True)\
            .load(f'abfss://bronzenetflix@naiyarstorage.dfs.core.windows.net/{var_src_folder}')
```

```python
print(var_target_folder)
print(var_src_folder)
```

```python
df.limit(10).display()
```

```python
df.write.format('delta')\
    .mode('append')\
        .option('path',f'abfss://silvernetflix@naiyarstorage.dfs.core.windows.net/{var_target_folder}')\
            .save()
```
