```python
%sql
CREATE SCHEMA netflix_catalog.net_schema
```

```python
checkpointLocation='abfss://silvernetflix@naiyarstorage.dfs.core.windows.net/checkpoint'

df=spark.readStream.format('cloudFiles')\
    .option('cloudFiles.format','csv')\
    .option('cloudFiles.schemaLocation',checkpointLocation)\
    .load('abfss://rawnetflix@naiyarstorage.dfs.core.windows.net/')
```

```python
df.writeStream\
  .option('checkpointLocation',checkpointLocation)\
    .trigger(once=True)\
      .start('abfss://bronzenetflix@naiyarstorage.dfs.core.windows.net/netflix_titles')
```
