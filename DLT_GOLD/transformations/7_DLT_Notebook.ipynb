```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from delta.tables import DeltaTable
```

```python
looktables_rule={
    "rule1":"show_id is NOT NULL"
}
```

```python
@dlt.table(
    name="gold_netflix_directors"
)
@dlt.expect_all_or_drop(looktables_rule)
def gold_netflix_category():
    df = spark.readStream.format('delta').load(
        'abfss://silvernetflix@naiyarstorage.dfs.core.windows.net/netflix_category'
    )
    return df
```
